```python
import pandas as pd
import seaborn as sns
sns.set(rc={'figure.figsize':[25,25]} , font_scale = 1.5 )
```

```python
cops = pd.read_csv("911.csv - 911.csv.csv")
cops
```

```python
cops.info()
```

# NUMERICAL UNIVARIATE PLOTS
=========================================================

```python
sns.displot(cops["hour"],color = "m",kde=True)
```

### most calls between 8am : 8pm

```python
sns.displot(cops["month"] , color = "m" , kde = True)
```

### most calls in January then July
### In October & November there are not many calls
### We can say that calls in the past three months are few compared to those before

```python
sns.displot(cops["Year"] , color = "m" , kde = True)
```

### No calls in 2015 compared to 2016, Almost all problems occurred in 2016

# CATEGORICAL UNIVARIATE PLOTS
=========================================================

```python
sns.countplot(data = cops , y = "twp" , palette = "icefire")
```

### Lower Merion is the most connected to 911 then Abington & Norristown are close in number
### Least contacted countries, in order: Lehigh County, Phila County, Green Lane

```python
sns.countplot(data = cops , y = "twp" , palette = "icefire" , hue = "Reason")
```

### Norristown is the most town has a EMS problem
### Lower Merion is the most town has a traffic problem and fire problem
### there are no calls from Lehigh County except EMS calls

```python
sns.countplot(data = cops , x = "Reason" , palette = "icefire")
```

### there are more than 50000 calls due to EMS and this is the most calls
### fire calls are the least one

```python
sns.countplot(data = cops , x = "Reason" , palette = "icefire" , hue = "day of week")
```

### Close fire results

### Tue & Wed are the most days that traffic problems happen
### Sun is the least day there were problems
